## 1. Review if else

```python
def bilangan(bil):
    if bil == 1:
        return "Satu"
    elif bil == 2:
        return "Dua"
    elif bil == 3:
        return "Tiga"
    elif bil == 4:
        return "Empat"
    elif bil == 5:
        return "Lima"
    elif bil == 6:
        return "Enam"
    elif bil == 7:
        return "Tujuh"
    elif bil == 8:
        return "Delapan"
    else:
        return "Sembilan"
```

```python
def sebutkan(angka):
    if angka < 10:
        return bilangan(angka)
    elif angka == 10:
        return "Sepuluh"
    elif angka == 11:
        return "Sebelas"
    elif angka > 11 and angka < 20:
        return bilangan(angka%10) + " belas"
    else:
        if angka%10 == 0:
            return bilangan(int(angka/10)) + " Puluh"
        else:
            return bilangan(int(angka/10)) + " Puluh "+ bilangan(angka%10)
```

```python
sebutkan(78)
```

```python
def klasifikasi(x):
    if x < 70:
        return 'Sekolah dodol'
    elif x < 75:
        return 'Sekolah lumayan'
    else:
        return 'Sekolah Puinter'
```

```python
def total_classification(total):
    if total < 250:
        return "sedikit"
    elif total < 300:
        return "lumayan"
    else:
        return "banyak"
```

# Matplotlib
## Plot

```python
x = [2000, 2012, 2013, 2024]
y = [70, 80, 87, 90]

plt.plot(x, y)
plt.xlabel("Tahun")
plt.ylabel("Nilai")
plt.show()
```

```python
x2 = [2001, 2013, 2015, 2020]
y2 = [70, 70, 77, 80]

plt.plot(x, y, label = 'Sekolah A')
plt.plot(x2, y2, label = 'Sekolah B')
plt.title("Rata-rata nilai ujian")
plt.xlabel("Tahun")
plt.ylabel("Nilai")
plt.legend()
plt.show()
```

```python
plt.scatter(x, y)

plt.title("Rata-rata nilai ujian")
plt.xlabel("Tahun")
plt.ylabel("Nilai")
plt.show()
```

```python
plt.scatter(x, y, label = 'Sekolah A', s = 250)
plt.scatter(x2, y2, label = 'Sekolah B')
plt.title("Rata-rata nilai ujian")
plt.xlabel("Tahun")
plt.ylabel("Nilai")
plt.legend()
plt.show()
```
