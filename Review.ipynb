## 1. Review if else

```python
def bilangan(bil):
    if bil == 1:
        return "Satu"
    elif bil == 2:
        return "Dua"
    elif bil == 3:
        return "Tiga"
    elif bil == 4:
        return "Empat"
    elif bil == 5:
        return "Lima"
    elif bil == 6:
        return "Enam"
    elif bil == 7:
        return "Tujuh"
    elif bil == 8:
        return "Delapan"
    else:
        return "Sembilan"
```

```python
def sebutkan(angka):
    if angka < 10:
        return bilangan(angka)
    elif angka == 10:
        return "Sepuluh"
    elif angka == 11:
        return "Sebelas"
    elif angka > 11 and angka < 20:
        return bilangan(angka%10) + " belas"
    else:
        if angka%10 == 0:
            return bilangan(int(angka/10)) + " Puluh"
        else:
            return bilangan(int(angka/10)) + " Puluh "+ bilangan(angka%10)
```

```python
sebutkan(78)
```
